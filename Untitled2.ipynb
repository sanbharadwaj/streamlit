```python
import pandas as pd
import json
```

```python
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 100)
```

```python
with open('b2.json', 'r') as file:
    # load the contents of the file into a variable
    data = json.load(file)
```

```python
print(type(data))
```

```python
dfo = pd.json_normalize(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][0]['originDestinations'][0]['segments'])
dfo['originIndex'] = 0
dfo.set_index("originIndex", inplace=True)

for i in range(1, len(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][0]['originDestinations'])):
    dfi = pd.json_normalize(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][0]['originDestinations'][i]['segments'])
    dfi['originIndex'] = i
    dfi.set_index("originIndex", inplace=True)
    dfo = pd.concat([dfo,dfi])
```

```python
dfr = pd.json_normalize(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][1]['originDestinations'][0]['segments'])
dfr['destinationIndex'] = 0
dfr.set_index("destinationIndex", inplace=True)

for i in range(1, len(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][1]['originDestinations'])):
    dfi = pd.json_normalize(data['searchCriteriaResponses'][0]['scheduleCriteriaResponse'][1]['originDestinations'][i]['segments'])
    dfi['destinationIndex'] = i
    dfi.set_index("destinationIndex", inplace=True)
    dfr = pd.concat([dfr,dfi])
```
